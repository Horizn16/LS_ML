```python
def numpy_boundary_traversal(matrix):
    rows, cols = matrix.shape
    result = []

    # Traverse the first row
    for j in range(cols):
        result.append(matrix[0, j])

    # Traverse the last column
    for i in range(1, rows):
        result.append(matrix[i, cols - 1])

    # Traverse the last row if there are more than one row
    if rows > 1:
        for j in range(cols - 2, -1, -1):
            result.append(matrix[rows - 1, j])

    # Traverse the first column if there are more than one column
    if cols > 1:
        for i in range(rows - 2, 0, -1):
            result.append(matrix[i, 0])

    return result
```

```python
def numpy_alternate_sort(array):
    sorted_arr = np.sort(array)
    result = []
    left, right = 0, len(sorted_arr) - 1
    while left <= right:
        result.append(sorted_arr[left])
        left += 1
        if left <= right:
            result.append(sorted_arr[right])
            right -= 1
    return np.array(result)
```

```python

# 2. Preprocessing with TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=300, lowercase=True, stop_words='english')
X = vectorizer.fit_transform(data['Feedback'])
y = data['Label']

# 3. Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Train Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# 5. Evaluate on test set
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

# 6. Function: text_preprocess_vectorize
def text_preprocess_vectorize(texts, vectorizer):
    """
    Vectorizes a list of text samples using a fitted TfidfVectorizer.
    
    Parameters:
    - texts: List of strings (text samples)
    - vectorizer: Fitted TfidfVectorizer
    
    Returns:
    - TF-IDF feature matrix
    """
    return vectorizer.transform(texts)

# Example use
example_texts = ["This product is amazing!", "Completely useless."]
example_features = text_preprocess_vectorize(example_texts, vectorizer)
example_preds = model.predict(example_features)

for text, pred in zip(example_texts, example_preds):
    print(f"Review: '{text}' → Sentiment: {pred}")
```
